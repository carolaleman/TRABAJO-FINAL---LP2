```python
#Importamos las librerías necesarias:
from bs4 import BeautifulSoup
import requests

entry_list = []
name_list = []
protein_list = []
gene_list = []
organism_list = []
status_list = []
nomgo_list = []
molfunc_list = []
molecular_2_list = []

#Realizamos un bucle para obtener todos los urls:
gen= ["robo1"]
for i in gen:
    
    url = "https://www.uniprot.org/uniprot/?query="+i+"+bos+taurus&sort=score"

    html = requests.get(url).text

    soup = BeautifulSoup(html, "lxml")
    
    #Si no hay resultado de la búsqueda, se omiten los siguientes pasos:
    empty = soup.find("div", {"id" : "noResultsMessage"})
    if empty is not None:
        entry = " "
        name = " "
        protein = " "
        gene = " "
        organism = " "
        status = " "
        nomgo = " "
        molfunc = " "
        molecular_2 = " "
    #Webscrapeamos lo pedido en el reporte:
    else:
        entry = soup.find("td", class_="entryID").text
        if entry is None:
            
            entry = " "

        url_2 = "https://www.uniprot.org/uniprot/"+entry
        
        html_2 = requests.get(url_2).text

        soup_2 = BeautifulSoup(html_2, "lxml")

        name = soup.find_all("td", class_= False)[1].text
        if name is None:
            name = " "
     
        protein = soup_2.find("div", {"id" : "content-protein"}).text
        if protein is None:
            protein = " "
        
        gene = soup_2.find("div", {"id" : "content-gene"}).text
        if gene is None:
            gene = " "
        
        organism = soup_2.find("div", {"id" : "content-organism"}).text
        if organism is None:
            gene = " "
        
        status = soup_2.find("a", class_="icon-uniprot reviewed-icon tooltipped")

        if status is None:
            status = "Unreviewed"
        else:
            status = "Reviewed"
        
        go_list = soup_2.find("ul", class_="noNumbering biological_process")
        if go_list is not None:
            for go in go_list:
                nomgo=go.a.text
        else:
            nomgo = " "
        
        funcion = soup_2.find('div', class_="section", id="function")
        
        keyword= funcion.find("table",class_="databaseTable")
        
        if keyword is None:

            molfunc = " "
            molecular_2= " "
            
        else: 
            key=keyword.find_all('tr')
        
            l=[]
            l1=[]
            for i in key:
                if i.find("td").text =="Molecular function":
                    clave = i.find_all("a")
                    a=""
                    for i in clave:
                        a+= i.text+", "
                    
                elif i.find("td").text == "Biological process":
                    clave = i.find_all("a")
                    b=""
                    for i in clave:
                        b+= i.text+", "
            if a is None:
                molfunc_list.append(" ")
            else:
                molfunc_list.append(a)
            if b is None:
                molecular_2_list.append(" ")
            else:
                molecular_2_list.append(b)
            
             
                     
            
            
    entry_list.append(entry)
    name_list.append(name)
    protein_list.append(protein)
    gene_list.append(gene)
    organism_list.append(organism)
    status_list.append(status)
    nomgo_list.append(nomgo)

print(gen)
print(entry_list) 
print(name_list) 
print(protein_list) 
print(gene_list) 
print(organism_list) 
print(status_list) 
print(nomgo_list) 
print(molfunc_list)
print(molecular_2_list)
```
