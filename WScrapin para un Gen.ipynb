```python
#Importamos las librerías necesarias:
from bs4 import BeautifulSoup
import requests
import re
entry_list = []
name_list = []
protein_list = []
gene_list = []
organism_list = []
status_list = []
nomgo_list = []
molfunc_list = []
molecular_2_list = []

#Realizamos un bucle para obtener todos los urls:
gen= ["Robo1"]
for i in gen:
    
    url = "https://www.uniprot.org/uniprot/?query="+i+"+bos+taurus&sort=score"

    html = requests.get(url).text

    soup = BeautifulSoup(html, "lxml")
    
    #Si no hay resultado de la búsqueda, se omiten los siguientes pasos:
    empty = soup.find("div", {"id" : "noResultsMessage"})
    if empty is not None:
        entry = " "
        name = " "
        protein = " "
        gene = " "
        organism = " "
        status = " "
        nomgo = " "
        molfunc = " "
        molecular_2 = " "
        #Webscrapeamos lo pedido en el reporte:
    else: 
        entry = soup.find("td", class_="entryID").text
        if entry is None:
            
            entry = " "

        url_2 = "https://www.uniprot.org/uniprot/"+entry
        
        html_2 = requests.get(url_2).text

        soup_2 = BeautifulSoup(html_2, "lxml")

        name = soup.find_all("td", class_= False)[1].text
        if name is None:
            name = " "
     
        protein = soup_2.find("div", {"id" : "content-protein"}).text
        if protein is None:
            protein = " "
        
        gene = soup_2.find("div", {"id" : "content-gene"}).text
        if gene is None:
            gene = " "
        
        organism = soup_2.find("div", {"id" : "content-organism"}).text
        if organism is None:
            gene = " "
        
        status = soup_2.find("a", class_="icon-uniprot reviewed-icon tooltipped")

        if status is None:
            status = "Unreviewed"
        else:
            status = "Reviewed"
        
        go_list = soup_2.find("ul", class_="noNumbering biological_process")
        if go_list is not None:
            for go in go_list:
                nomgo=go.a.text
                nomgo_list.append(nomgo)
                
        else:
            nomgo = " "
        
        funcion = soup_2.find('div', class_="section", id="function")  #Buscaremos en el html la clase "section" que tenga 
                                                                       #id "function"
        keyword= funcion.find("table",class_="databaseTable")          #Dentro de esto mismo buscaremos la classe 
                                                                       #"databaseTable" que es donde se ecuentra almacenada 
        if keyword is None:                                            #la información

            molfunc = " "                                              #En caso de que keword no exista se imprimirán 
            molecular_2= " "                                           #espacios en blanco
             
        else:                                                          #En caso de que keywod exista se buscaran
            key=keyword.find_all('tr')                                 #las etiquetas tr
                                      
            for i in key:                                              #Para cada elemento en key
                if i.find("td").text =="Molecular function":           #Busaremos la etiqueta "td" que diga "Molecular function"
                    clave = i.find_all("a")                            #y dentro de esta buscaremos todas las etiquetas "a"
                    a=""
                    for i in clave:                                    #para cada elemento de clave
                        a+= i.text+", "                                #imprimiremos el texto mas una "," que lo separara del
                                                                       #de lo siguientes elementos en caso de que existan
                elif i.find("td").text == "Biological process":        #Igual que en el caso anterior pero ahora buscaremos
                    clave = i.find_all("a")                            #etiqueta "td" que diga Biological process
                    b=""
                    for i in clave:
                        b+= i.text+", "
            if a is None:                                               #En caso de que "a" no exista 
                molfunc_list.append(" ")                                #se imprimirá un espacio en blanco
            else:
                molfunc_list.append(a)                                  #En caso contrario se agregará a la lista
            if b is None:
                molecular_2_list.append(" ")                            #Igualemente para "b"
            else:
                molecular_2_list.append(b)
            
             
                     
            
#Agregaremos a las listas correspondientes cada dato imprimido             
    entry_list.append(entry)
    name_list.append(name)
    protein_list.append(protein)
    gene_list.append(gene)
    organism_list.append(organism)
    status_list.append(status)
    

#imprimiremos las listas resultantes
print(entry_list) 
print(name_list) 
print(protein_list) 
print(gene_list) 
print(organism_list) 
print(status_list) 
print(nomgo_list) 
print(molfunc_list)
print(molecular_2_list)
```
